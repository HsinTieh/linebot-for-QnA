```
import jieba
jieba.set_dictionary('dict.txt')
with open('stops.txt', 'r', encoding='utf8') as f:
  stops=f.read().split('\n')
```

```
#測試
print([t for t in jieba.cut("今天天氣很好，很適合出去打球")])

record_num = int(qna_.describe().ix[0,0])
print(record_num,qna_.ix[0,:])
```

```
segments = []

for i in range(record_num):
  splitedStr=''
  content=qna_.ix[i,:]
  words = jieba.cut(content['question'])
  for word in words:
    if word not in stops:
     # segments.append({'word':word,'count':1})
      splitedStr +=word+' '
  segments.append(splitedStr)
print (segments[1])
```

```
import os
workspace_path = "./"
with open(os.path.join(workspace_path, 'train.data'),'w', encoding='utf-8') as f:
  f.write('\n'.join(segments))
!ls
```

```
#資料分割 for linebot
split =[]
count=0
w11=[]
w22=[]
w33=[]
w44=[]
w55=[]
ww=[]


for item in segments:
  split=item.split()
  w1=''
  w2=''
  w3=''
  w4=''
  w5=''
  w=''
  for s in split:
    if len(s)==1:
      w1+=s+' '
      
    elif len(s)==2:
      w2+=s+' '
      
    elif len(s)==3:
      w3+=s+' '
      
    elif len(s)==4:
      w4+=s+' '
      
    elif len(s)==5:
      w5+=s+' '
      
    else :
      w+=s
  w11.append([count,w1])
  w22.append([count,w2])
  w33.append([count,w3])
  w44.append([count,w4])
  w55.append([count,w5])
  ww.append([count,w])
  count+=1
  
```

```
with open ('w1.csv','w',newline='') as csvFile:
  writer =csv.writer(csvFile)
  #title
  #writer.writerow(['id','word'])
  
  #data
  for w in w11:
    writer.writerow([w[0],w[1]])

with open ('w2.csv','w',newline='') as csvFile:
  writer =csv.writer(csvFile)
  #title
  #writer.writerow(['id','word'])
  
  #data
  for w in w22:
    writer.writerow([w[0],w[1]])
    
with open ('w3.csv','w',newline='') as csvFile:
  writer =csv.writer(csvFile)
  #title
  #writer.writerow(['id','word'])
  
  #data
  for w in w33:
    writer.writerow([w[0],w[1]])

with open ('w4.csv','w',newline='') as csvFile:
  writer =csv.writer(csvFile)
  #title
  #writer.writerow(['id','word'])
  
  #data
  for w in w44:
    writer.writerow([w[0],w[1]])
    
with open ('w5.csv','w',newline='') as csvFile:
  writer =csv.writer(csvFile)
  #title
  #writer.writerow(['id','word'])
  
  #data
  for w in w55:
    writer.writerow([w[0],w[1]])
    
with open ('w6.csv','w',newline='') as csvFile:
  writer =csv.writer(csvFile)
  #title
  #writer.writerow(['id','word'])
  
  #data
  for w in ww:
    writer.writerow([w[0],w[1]])
```

#**5.測試(要整理好寫在app.py)**

```
test=''
mes=[]
words = jieba.cut("臺北市及新北市之敬老卡、愛心卡、愛心陪伴卡搭乘臺北捷運、淡海輕軌及臺北地區聯營公車等運具之扣點/扣款方式為何？")
for word in words:
  if word not in stops:
    test +=word+' '
    mes.append(word)
print(test)
```

```

pro_qna_list=[]
with open('processed.csv', 'r') as pro_qna:
  for line in pro_qna.readlines():
    pro_qna_cut=line.strip()
    pro_qna_list.append(pro_qna_cut.split(','))
```

**計算詢問的語句被切割成幾個字**<br>
[ *0~5* ] 表示 [ 第一個字~第六個字(被切成6以上的字) ]<br>
假設enable[1] = 0<br>
表示詢問語句並沒有被切成二個字

```
enable=[0,0,0,0,0,0]
for m in mes:
  if len(m)==1:
    enable[0]=1
  elif len(m)==2:
    enable[1]=1
  elif len(m)==3:
    enable[2]=1
  elif len(m)==4:
    enable[3]=1
  elif len(m)==5:
    enable[4]=1
  else :
    enable[5]=1
```

```
#for linebot use
def cal_index(i):
  r_list=[]
  findIndex=[] 
 
  #計算問句在字典(1~6)出現的次數
  if enable[i]==1:
    #讀取字典檔
    with open('w'+str(i+1)+'.csv', 'r') as r:
      for line in r.readlines():
        r_cut=line.strip()
        if r_cut=='':
          r_cut='error'
        r_list.append(r_cut.split(','))
    
      #計算出現次數
      terms=[]
      for term in r_list: 
        terms.append(term[1].split())
      for term in terms:
        count=0
        for t in term:
          for word in mes:
            if t==word:
              count+=1
              #print(t,word)
        findIndex.append(count)
  return findIndex
```

```
#output answer
finalIndex=[]
addweight=[0.5,1,2,1.5,1.5]
for i in range(len(enable)):
  findIndex= cal_index(i)
  #檢查findIndex是否存在
  if findIndex:
    #檢查finalIndex是否存在
    if finalIndex:
      #print(i,findIndex)
      findIndex=[findIndex[j]*addweight[i] for j in range(len(findIndex))]
      finalIndex=[finalIndex[j]+findIndex[j] for j in range(len(finalIndex))]
    else:
      #print(i,findIndex)
      finalIndex=([findIndex[j]*addweight[i] for j in range(len(findIndex))])

#print('fin',finalIndex)
maxindex=finalIndex.index(max(finalIndex))
print(maxindex,finalIndex[maxindex] )

#+1是因為 processing有欄位名稱
print(pro_qna_list[maxindex+1][0],pro_qna_list[maxindex+1][1])
```
